```python
import json
```

```python
f = open('../data/types.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)
```

```python
data
```

```python
relations = data['relations']
relations
```

```python
relations_dict = {}
count = 0
for relation, relation_dict in relations.items():
    relations_dict[relation_dict['verbose']] = int(relation[1:])
    count += 1
```

```python
relations_dict
```

```python
with open("../data/relations_dict.json", "w") as outfile:
    json.dump(relations_dict, outfile)
```
